Create a Data:

```python
names = [fake.name() for _ in range(250)]
data={
    "Student_Id" :np.arange(1,251),
    "Name" :np.random.choice(names, 250),
    "Gender" :np.random.choice(["male","female"],250),
    "Department" :np.random.choice(["HR","IT","Finance","Marketing"],250),
    "Year":np.random.randint(2025,2028,250),
    "Maths":np.random.randint(0,100,250),
    "Science":np.random.randint(0,100,250),
    "English":np.random.randint(0,100,250),
    "Attendance":np.random.randint(50,100,250),
    "Internal Marks":np.random.randint(0,25,250),
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3
```

```python
df["Grade"] = df["Average"].apply(lambda x: "A" if x>=85 else "B" if x>=70 else "C" if x>=50 else "Fail")
```

```python
print("\nTop Students:\n", df.nlargest(5, "Average")[["Name","Average","Grade"]])
```

Bar Chart

```python
plt.scatter(df["Attendance"], df["Average"])
plt.xlabel("Attendance")
plt.ylabel("Average Marks")
plt.title("Attendance vs Performance")
plt.show()
```

Pie Chart

```python
plt.figure()
df["Gender"].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

Histogram

```python
plt.figure()
plt.hist(df["Average"], bins=10)
plt.title("Distribution of Average Marks")
plt.xlabel("Average Marks")
plt.ylabel("Frequency")
plt.show()
```

Scatter Plot

```python
plt.figure()
plt.scatter(df["Attendance"], df["Average"])
plt.title("Attendance vs Average Marks")
plt.xlabel("Attendance")
plt.ylabel("Average Marks")
plt.show()
```
